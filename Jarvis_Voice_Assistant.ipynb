Jarvis Voice Assistant

```python
# pip install pyttsx3 speechrecognition wikipedia requests
import datetime
import random
import webbrowser
import os

# Text-to-speech (uncomment for actual voice)
# import pyttsx3
# engine = pyttsx3.init()
# def speak(text): engine.say(text); engine.runAndWait()

def speak(text):
    print(f"🤖 JARVIS: {text}")

def greet():
    hour = datetime.datetime.now().hour
    if 6 <= hour < 12: greeting = "Good morning"
    elif 12 <= hour < 18: greeting = "Good afternoon"
    else: greeting = "Good evening"
    speak(f"{greeting}! I'm Jarvis, your personal assistant. How can I help?")

def process_command(command):
    command = command.lower()
    if 'time' in command:
        now = datetime.datetime.now().strftime("%I:%M %p")
        speak(f"The current time is {now}")
    elif 'date' in command:
        today = datetime.datetime.now().strftime("%B %d, %Y")
        speak(f"Today is {today}")
    elif 'youtube' in command:
        speak("Opening YouTube")
        webbrowser.open("https://youtube.com")
    elif 'google' in command:
        query = command.replace('google', '').strip()
        if query:
            speak(f"Searching Google for {query}")
            webbrowser.open(f"https://www.google.com/search?q={query}")
    elif 'joke' in command:
        jokes = [
            "Why do programmers prefer dark mode? Because light attracts bugs!",
            "Why did the Python programmer wear glasses? Because he couldn't C!",
            "A SQL query walks into a bar, walks up to two tables and asks: Can I join you?",
        ]
        speak(random.choice(jokes))
    elif 'weather' in command:
        speak("You'd need an API key for live weather. Try OpenWeatherMap!")
    elif 'calculate' in command or 'math' in command:
        speak("For calculations, please provide an expression!")
    elif 'exit' in command or 'bye' in command:
        speak("Goodbye! Have a great day!")
        return False
    else:
        speak(f"I heard: '{command}'. I'm still learning. Try: time, date, joke, or google something.")
    return True

# Demo mode
greet()
test_commands = ["what time is it", "tell me a joke", "what is today's date", "bye"]
print()
for cmd in test_commands:
    print(f"👤 User: {cmd}")
    if not process_command(cmd): break
    print()
```
